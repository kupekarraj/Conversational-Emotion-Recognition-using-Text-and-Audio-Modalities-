```python
#IMPORTING THE REQUIRED LIBRARIES 
import pandas as pd 
import numpy as np
from matplotlib import pyplot
import seaborn as sns
```

```python
#Importing the data for training,testing and validation 
train_text=pd.read_csv(r"C:\Users\kupekarraj\Desktop\Thesis_Dataset\MELD.Raw\train_sent_emo.csv",encoding='cp1252')
test_text=pd.read_csv(r"C:\Users\kupekarraj\Desktop\Thesis_Dataset\MELD.Raw\test_sent_emo.csv")
val_text=pd.read_csv(r"C:\Users\kupekarraj\Desktop\Thesis_Dataset\MELD.Raw\dev_sent_emo.csv")
```

```python
#importing the required libraries for pre processing 
import nltk
nltk.download('punkt')
import string
import re
```

```python
#defining a function for cleaning the text 
def cleantext(x):
    text=re.sub("[^a-zA-Z]"," ",x)
    text=text.lower()
    text=text.split()
    text=" ".join(text)
    return(text)
```

```python
#Applying the cleantext function for cleaning the textual data in train, test and validation data
train_text["Utterance"]= train_text["Utterance"].apply(lambda x : cleantext(x))
test_text["Utterance"]=test_text["Utterance"].apply(lambda x : cleantext(x))
val_text["Utterance"]=val_text["Utterance"].apply(lambda x : cleantext(x))
```

```python
#Pre-processed text utterance
train_text["Utterance"][1]
```

```python
#Fetching the utterance and emotion values and forming the dependent and independent variables
x_train=train_text["Utterance"].values
Y_train=train_text['Emotion'].values
x_test=test_text["Utterance"].values
Y_test=test_text["Emotion"].values
x_val=val_text["Utterance"].values
Y_val=val_text["Emotion"].values
```

```python
#importing the required libraries for keras pre-processing 
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
NUM_WORDS=10000
tokenizer = Tokenizer(num_words=NUM_WORDS,filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n\'',
                      lower=True )
tokenizer.fit_on_texts(x_train)
sequences_train = tokenizer.texts_to_sequences(x_train)#creating tokenized sequences of pre-processed text 
sequences_valid=tokenizer.texts_to_sequences(x_val)#creating tokenized sequences of pre-processed text 
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
#initializing the maximum lenght of a sentence
max_len=(max(len(s.split()) for s in train_text["Utterance"]))
print(max_len)
```

```python
#Padding the sequences to an array of same length 
X_train = pad_sequences(sequences_train)
X_val = pad_sequences(sequences_valid,maxlen=X_train.shape[1])
Y_train = to_categorical(Y_train)#creating an array of length having a one hot encoded emotion labels
Y_val = to_categorical(Y_val)#creating an array of length having a one hot encoded emotion labels
print('Shape of X train and X validation tensor:', X_train.shape,X_val.shape)
print('Shape of Y train and Y validation tensor:', Y_train.shape,Y_val.shape)
```

***Building the Word2Vec Embedding Model***

```python
#importing the required libraries for building word2vec model
from gensim.utils import simple_preprocess
import gensim
from gensim.models import Word2Vec
import time
```

```python
#Tokenize the text column to get the new column 'tokenized_utterance'
token_utterance= [simple_preprocess(line, deacc=True) for line in train_text['Utterance']] 
token_utterance[1]
```

```python
#Creating the model and setting values for the various parameters
num_features = 300  # Word vector dimensionality
min_word_count = 0 # Minimum word count
num_workers = 4     # Number of parallel threads
context = 10        # Context window size
downsampling = 1e-3 # (0.001) Downsample setting for frequent words

# Initializing the train model
from gensim.models import word2vec
print("Training model....")
model = word2vec.Word2Vec(token_utterance,\
                          workers=num_workers,\
                          size=num_features,\
                          min_count=min_word_count,\
                          window=context,
                          sample=downsampling)

# To make the model memory efficient
model.init_sims(replace=True)
print(model)
```

```python
#Printing word vectors for few words and printing vocab list
print(model.wv.most_similar("man"))
words = list(model.wv.vocab)
print((words))
```

```python
#Saving the model in both .txt and .bin extension for further use
model.save("model.txt")
model.save("model.bin")
```

```python
#Loading the save word2vec file 
file="model.bin"
model_vec = Word2Vec.load(file)
print("Word2Vec Model Statistics")
print(model_vec)
```

```python
#Creating the word2vec for the training text
EMBEDDING_DIM=300
vocabulary_size=min(len(word_index)+1,NUM_WORDS)
embedding_matrix = np.zeros((vocabulary_size, EMBEDDING_DIM))
for word, i in word_index.items():
    if i>=NUM_WORDS:
        continue
    try:
        embedding_vector = model_vec[word]
        embedding_matrix[i] = embedding_vector
    except KeyError:
        embedding_matrix[i]=np.random.normal(0,np.sqrt(0.25),EMBEDDING_DIM)

del(model_vec)
```

```python
#Importing the required libraries for building the BiLSTM model
from keras.layers import Dropout, Dense, Embedding, LSTM, Bidirectional, Input, Conv1D
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam

sequence_length = X_train.shape[1]

#Building the Bi-LSTM model with Word2Vec embedding layer
model_LSTM= Sequential()
model_LSTM.add(Embedding(vocabulary_size,
                                EMBEDDING_DIM,
                                weights=[embedding_matrix],
                                input_length= sequence_length,
                                trainable=True))
model_LSTM.add(Bidirectional(LSTM(64,activation="relu",recurrent_dropout=0.2,kernel_initializer ="uniform",return_sequences=True)))
model_LSTM.add(LSTM(64,activation="relu",recurrent_dropout=0.2)) 
model_LSTM.add(Dropout(0.5))
model_LSTM.add(Dense(128,activation="relu"))
model_LSTM.add(Dense(7,activation="softmax"))

#compiling the model using optimizer and loss fucntion
model_LSTM.compile(loss='categorical_crossentropy', optimizer= "rmsprop", metrics=['accuracy'])
callbacks = [EarlyStopping(monitor='accuracy')]

#printing the summary of the model
model_LSTM.summary()
```

```python
#Training the build model on training data
start_time = time.time()
callbacks = [EarlyStopping(monitor='accuracy')]
model_fit = model_LSTM.fit(X_train, Y_train, batch_size=50, epochs=20, validation_data=(X_val, Y_val),callbacks=callbacks)

final_time = (time.time()- start_time)/60
```

```python
#Plotting the graph of model accuracy versus validation accuracy
model_loss = pd.DataFrame(model_LSTM.history.history)
model_loss.head()
model_loss[['accuracy','val_accuracy']].plot(ylim=[0,1]);
```

```python
#Preparing the testing data for evaluation and predict using the trained model
sequences_test=tokenizer.texts_to_sequences(x_test)
X_test = pad_sequences(sequences_test,maxlen=X_train.shape[1])
Y_pred=model_LSTM.predict(X_test)
Y_test=to_categorical(Y_test)
```

```python
from sklearn.metrics import confusion_matrix

#defining a function for confusion matrix using a heatmap
def show_confusionmatrix(y_actual, y_predicted, title, figsize=(5,4)):
    confmat= confusion_matrix(y_actual, y_predicted, labels=np.unique(y_actual))
    confmat_sum = np.sum(confmat, axis=1, keepdims=True)
    confmat_percent = confmat / confmat_sum.astype(float) * 100
    heatmp = np.empty_like(confmat).astype(str)
    nrows, ncols = confmat.shape
    for i in range(nrows):
        for j in range(ncols):
            c = confmat[i, j]
            p = confmat_percent[i, j]
            if i == j:
                s = confmat_sum[i]
                heatmp[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            else:
                heatmp[i, j] = '%.1f%%\n%d' % (p, c)
    confmat = pd.DataFrame(confmat, index=np.unique(y_actual), columns=np.unique(y_actual))
    confmat.index.name = 'Actual'
    confmat.columns.name = 'Predicted'
    fig, ax = pyplot.subplots(figsize=figsize)
    pyplot.title(title)
    sns.heatmap(confmat, annot=heatmp, fmt='', cmap= "Wistia",ax=ax,linewidths=.5)
```

```python
#genrating the confusion matrix
show_confusionmatrix(Y_test.argmax(axis=1),Y_pred.argmax(axis=1), 'Confusion matrix', figsize=(10,8))
print("The confusion matrix is: ")
confusion_matrix(Y_test.argmax(axis=1), Y_pred.argmax(axis=1), labels=np.unique(Y_test.argmax(axis=1)))
```

***Building a CNN model

```python
from keras.layers import Embedding
sequence_length = X_train.shape[1]
embedding_layer = Embedding(vocabulary_size,
                            EMBEDDING_DIM,
                            weights=[embedding_matrix],
                            trainable=True)
```

```python
from keras.layers import Dense, Input, GlobalMaxPooling1D
from keras.layers import Conv1D, MaxPooling1D, Embedding
from keras.models import Model
from keras.layers import Input, Dense, Embedding, Conv2D, MaxPooling2D, Dropout,concatenate
from keras.layers.core import Reshape, Flatten
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam
from keras.models import Model
from keras import regularizers
sequence_length = X_train.shape[1]
filter_sizes = [3,4,5]
num_filters = 100
drop = 0.5

inputs = Input(shape=(sequence_length,))
embedding = embedding_layer(inputs)
reshape = Reshape((sequence_length,EMBEDDING_DIM,1))(embedding)

conv_0 = Conv2D(num_filters, (filter_sizes[0], EMBEDDING_DIM),activation='relu',kernel_regularizer=regularizers.l2(0.01))(reshape)
conv_1 = Conv2D(num_filters, (filter_sizes[1], EMBEDDING_DIM),activation='relu',kernel_regularizer=regularizers.l2(0.01))(reshape)
conv_2 = Conv2D(num_filters, (filter_sizes[2], EMBEDDING_DIM),activation='relu',kernel_regularizer=regularizers.l2(0.01))(reshape)

maxpool_0 = MaxPooling2D((sequence_length - filter_sizes[0] + 1, 1), strides=(1,1))(conv_0)
maxpool_1 = MaxPooling2D((sequence_length - filter_sizes[1] + 1, 1), strides=(1,1))(conv_1)
maxpool_2 = MaxPooling2D((sequence_length - filter_sizes[2] + 1, 1), strides=(1,1))(conv_2)
merged_tensor = concatenate([maxpool_0, maxpool_1, maxpool_2], axis=1)
flatten = Flatten()(merged_tensor)
reshape = Reshape((3*num_filters,))(flatten)
dropout = Dropout(drop)(flatten)
output = Dense(units=7, activation='softmax',kernel_regularizer=regularizers.l2(0.01))(dropout)

# this creates a model that includes
model_CNN= Model(inputs, output)

model_CNN.summary()
```

```python
#Compiling and training the model on training data
model_CNN.compile(loss='categorical_crossentropy',
              optimizer="rmsprop",
              metrics=['accuracy'])
callbacks = [EarlyStopping(monitor='accuracy')]
model_CNN_fit=model_CNN.fit(X_train, Y_train, batch_size=100, epochs=50, verbose=1, validation_data=(X_val, Y_val), callbacks=callbacks)
```

```python
#Plotting the graph of model accuracy versus validation accuracy
model_loss_CNN = pd.DataFrame(model_CNN.history.history)
model_loss_CNN.head()
model_loss_CNN[['accuracy','val_accuracy']].plot(ylim=[0,1]);
```

```python
#Predicting on CNN model 
Y_pred_CNN=model_CNN.predict(X_test)
```

```python
#genrating the confusion matrix
show_confusionmatrix(Y_test.argmax(axis=1),Y_pred_CNN.argmax(axis=1), 'Confusion matrix', figsize=(10,8))
print("The confusion matrix is: ")
confusion_matrix(Y_test.argmax(axis=1), Y_pred_CNN.argmax(axis=1), labels=np.unique(Y_test.argmax(axis=1)))
```
