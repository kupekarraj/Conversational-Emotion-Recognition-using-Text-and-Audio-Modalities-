```python
#Importing the required libraries 
import numpy as np
import pandas as pd
import seaborn as sns
import librosa 
import librosa.display
import matplotlib.pyplot as plt
import glob
import os
```

```python
#Pre-processing of a sample wav file 
file="C:\\Users\\kupekarraj\\Desktop\\MELD.Raw\\MELD.Raw\\dev\\dev_wav\\dia0_utt0.wav"
```

```python
#Load audio file with Librosa
signal, sample_rate = librosa.load(file, sr=22050)
```

```python
# WAVEFORM
#Display waveform
librosa.display.waveplot(signal, sample_rate, alpha=0.4)
plt.xlabel("Time's")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

```python
#Plotting for FFT -> power spectrum
fft = np.fft.fft(signal)

#Calculate absolute values on complex numbers to get magnitude
spectrum = np.abs(fft)

#Create frequency variable
f = np.linspace(0, sample_rate, len(spectrum))

#considering only the half of the spectrum and frequency
left_spectrum = spectrum[:int(len(spectrum)/2)]
left_f = f[:int(len(spectrum)/2)]

#Plot spectrum
plt.plot(left_f, left_spectrum, alpha=0.4)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Power spectrum")
```

```python
#Plot for STFT -> spectrogram
hop_length = 512 # in num. of samples
n_fft = 2048 # window in num. of samples

#Calculate duration hop length and window in seconds
hop_length_duration = float(hop_length)/sample_rate
n_fft_duration = float(n_fft)/sample_rate

print("STFT hop length duration is: {}s".format(hop_length_duration))
print("STFT window duration is: {}s".format(n_fft_duration))

#Perform stft
stft = librosa.stft(signal, n_fft=n_fft, hop_length=hop_length)

#Calculate absolute values on complex numbers to get magnitude
spectrogram = np.abs(stft)

#Display spectrogram
librosa.display.specshow(spectrogram, sr=sample_rate, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.title("Spectrogram")
```

```python
#Apply logarithm to cast amplitude to Decibels
log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sample_rate, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format="%+2.0f dB")
```

```python
#Extracting the dependent and independent variables 
x_ind=mfcc_d.iloc[:,:-1]
y_denp=mfcc_d.iloc[:,-1]

#Converting them into a numpy array
X = np.array(x_ind)
Y = np.array(y_denp)
```

```python
#creating an array of length having a one hot encoded emotion labels
from keras.utils import to_categorical
Y = to_categorical(Y)
```

```python
#Spliting the data into training and testing 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#Reshaping the inputs of independent variables
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],1)



print(X_train.shape)
print(X_test.shape)
```

```python
#Building an LSTM model with the required libraries
from keras.layers import Dropout, Dense, Embedding, LSTM
from keras.optimizers import SGD
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam
```

```python
#Building the model 
input_shape=(X_train.shape[1],1)

model_LSTM= Sequential()
model_LSTM.add(LSTM(64,input_shape=input_shape,activation="relu",recurrent_dropout=0.2,return_sequences=True))
model_LSTM.add(LSTM(64,activation="relu",recurrent_dropout=0.2))
model_LSTM.add(Dropout(0.5))
model_LSTM.add(Dense(128,activation="relu"))
model_LSTM.add(Dense(7,activation="softmax"))
```

```python
##compiling the model using optimizer and loss fucntion
opt=Adam(lr=0.01, decay=1e-6)
optimizer = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model_LSTM.compile(loss='categorical_crossentropy', optimizer= optimizer, metrics=['accuracy'])
callbacks = [EarlyStopping(monitor='accuracy')]

#printing the summary of the model
model_LSTM.summary()
```

```python
import time
start_time = time.time()
callbacks = [EarlyStopping(monitor='accuracy')]
model_fit = model_LSTM.fit(X_train, Y_train, batch_size=100, validation_split= 0.2, epochs=20,callbacks=callbacks)

final_time = (time.time()- start_time)/60
```

```python
#Plotting the graph of model accuracy versus validation accuracy
model_loss = pd.DataFrame(model_LSTM.history.history)
model_loss.head()
model_loss[['accuracy','val_accuracy']].plot(ylim=[0,1]);
```

```python
#Preparing the testing data for evaluation and predict using the trained model
Y_pred=model_LSTM.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

#defining a function for confusion matrix using a heatmap
def show_confusionmatrix(y_actual, y_predicted, title, figsize=(5,4)):
    confmat= confusion_matrix(y_actual, y_predicted, labels=np.unique(y_actual))
    confmat_sum = np.sum(confmat, axis=1, keepdims=True)
    confmat_percent = confmat / confmat_sum.astype(float) * 100
    heatmp = np.empty_like(confmat).astype(str)
    nrows, ncols = confmat.shape
    for i in range(nrows):
        for j in range(ncols):
            c = confmat[i, j]
            p = confmat_percent[i, j]
            if i == j:
                s = confmat_sum[i]
                heatmp[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            else:
                heatmp[i, j] = '%.1f%%\n%d' % (p, c)
    confmat = pd.DataFrame(confmat, index=np.unique(y_actual), columns=np.unique(y_actual))
    confmat.index.name = 'Actual'
    confmat.columns.name = 'Predicted'
    fig, ax = pyplot.subplots(figsize=figsize)
    pyplot.title(title)
    sns.heatmap(confmat, annot=heatmp, fmt='', cmap= "Wistia",ax=ax,linewidths=.5)
```
